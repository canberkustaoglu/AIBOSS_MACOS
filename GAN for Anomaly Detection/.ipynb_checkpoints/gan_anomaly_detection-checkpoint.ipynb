```python
import torch
import torch.nn as nn
import torch.nn.functional as F  
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from time import time
from tqdm import tqdm

# Set random seed for reproducibility
manualSeed = 2019
torch.manual_seed(manualSeed)
np.random.seed(manualSeed)

device = torch.device("cpu")
```

```python
# Import data
data = pd.read_csv('flame.csv', header=None)
data = data.values
data = torch.tensor(data, device=device)
```

```python
# Define hyperparameters
G_D_num = 10
G_group = {}
D_group = {}
z_avg = 0
BATCH_SIZE = 16
LR_G = 0.002  # Learning rate for generator
LR_D = 0.002  # Learning rate for discriminator
random_neuron = 100  # Number of neurons as input for generator
components = 2  # Total number of points that G generates
critic_num = 3
D_loss_history = []
G_loss_history = []
```

```python
# Define network classes
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.fc1 = nn.Linear(random_neuron, 128)
        self.fc2 = nn.Linear(128, 128)
        self.fc3 = nn.Linear(128, components)

    def forward(self, input):
        x = F.relu(self.fc1(input))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.fc1 = nn.Linear(components, 128)
        self.fc2 = nn.Linear(128, 128)
        self.fc3 = nn.Linear(128, 1)

    def forward(self, input):
        x = F.relu(self.fc1(input))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        x = torch.sigmoid(x)
        return x
```

```python
torch.autograd.set_detect_anomaly(True)  # Enable anomaly detection mode
```

```python
# Training
for iter in range(G_D_num):

    begin = time()

    D = Discriminator().to(device)
    G = Generator().to(device)

    # Optimization
    opt_D = torch.optim.Adam(D.parameters(), lr=LR_D)
    opt_G = torch.optim.Adam(G.parameters(), lr=LR_G)

    plt.ion()

    for step in tqdm(range(10000)):

        for _ in range(critic_num):

            # Random samples of real data
            idx = np.random.choice(len(data), BATCH_SIZE)
            selected_real = data[idx].float()

            # Random noises
            G_noise = torch.randn(BATCH_SIZE, random_neuron)
            G_data = G(G_noise)

            prob_real = D(selected_real)  # D tries to increase this probability
            prob_fake = D(G_data)  # D tries to decrease this probability

            D_loss = -torch.mean(torch.log(prob_real + 1e-9) + torch.log(1. - prob_fake + 1e-9))
            G_loss = torch.mean(torch.log(1. - prob_fake + 1e-9))

            opt_D.zero_grad()
            D_loss.backward(retain_graph=True)  # Reusing computational graph
            opt_D.step()

        # Calculate G_loss with respect to G_data
        prob_fake = D(G_data)
        G_loss = -torch.mean(torch.log(prob_fake + 1e-9))  # Update G_loss here

        opt_G.zero_grad()
        G_loss.backward()
        opt_G.step()

        if step % 5000 == 0 and step > 0:
            plt.cla()
            plt.plot(selected_real.numpy()[:, 0], selected_real.numpy()[:, 1], '.b', label='Real')
            plt.plot(G_data.detach().numpy()[:, 0], G_data.detach().numpy()[:, 1], '.r', label='Fake')
            plt.legend(loc='upper right', fontsize=10)
            plt.show()

    end = time()
    print(str(iter + 1) + '/' + str(G_D_num) + ' is done! Time is {0:0.2f} seconds;'.format((end - begin)))

    plt.ioff()
    plt.show()

    # Visualize the background
    x = np.arange(round(min(data.numpy()[:, 0])) - 10, round(max(data.numpy()[:, 0])) + 10, 0.1)
    y = np.arange(round(min(data.numpy()[:, 1])) - 10, round(max(data.numpy()[:, 1])) + 10, 0.1)
    xx, yy = np.meshgrid(x, y, sparse=True)

    a = xx + yy - yy
    b = xx + yy - xx
    mem = np.concatenate([
        a.reshape([a.shape[0], a.shape[1], 1]),
        b.reshape([a.shape[0], a.shape[1], 1])
    ], axis=2)

    z = D(torch.from_numpy(mem).float()).detach().numpy()
    plt.plot(selected_real.numpy()[:, 0], selected_real.numpy()[:, 1], '.b', label='Real')
    plt.plot(G_data.detach().numpy()[:, 0], G_data.detach().numpy()[:, 1], '.r', label='Fake')
    h = plt.contourf(x, y, z.reshape(z.shape[0], z.shape[1]), 20, cmap='Spectral')
    plt.colorbar()
    plt.show()

    z_avg = z_avg + z / G_D_num

    G_group['G' + str(iter)] = G
    D_group['D' + str(iter)] = D
```

## gan-2d-stack-gaussian modellemesi için çalışmalar gerçekleştirelim.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from time import time
from tqdm import tqdm

# Set random seed for reproducibility
manualSeed = 2019
torch.manual_seed(manualSeed)
np.random.seed(manualSeed)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# import data
data = pd.read_csv('toy.csv', header=None)
data = data.values
data = torch.tensor(data, device=device)

# define hyper-parameters
G_D_num = 10
G_group = {}
D_group = {}
z_avg = 0
BATCH_SIZE = 16
LR_G = 0.001  # learning rate for generator
LR_D = 0.001  # learning rate for discriminator
random_neuron = 100  # number of neurons as input of generator
components = 2  # total number of points that G generated
critic_num = 3
D_loss_history = []
G_loss_history = []

# define network class
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()
        self.main = nn.Sequential(
            nn.Linear(random_neuron, 128),
            nn.ReLU(True),
            nn.Linear(128, 128),
            nn.ReLU(True),
            nn.Linear(128, components)
        )

    def forward(self, input):
        return self.main(input)


class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()
        self.main = nn.Sequential(
            nn.Linear(components, 128),
            nn.ReLU(True),
            nn.Linear(128, 128),
            nn.ReLU(True),
            nn.Linear(128, 1),
            nn.Sigmoid()
        )

    def forward(self, input):
        return self.main(input)


# training
for iter in range(G_D_num):

    begin = time()

    D = Discriminator().to(device)
    G = Generator().to(device)

    # optimization
    opt_D = torch.optim.Adam(D.parameters(), lr=LR_D)
    opt_G = torch.optim.Adam(G.parameters(), lr=LR_G)

    plt.ion()

    
    
    for step in tqdm(range(10000)):

        for _ in range(critic_num):

            # random samples of real data
            idx = np.random.choice(len(data), BATCH_SIZE)
            selected_real = data[idx].clone().detach().to(device).float()

            # random noises
            G_noise = torch.randn(BATCH_SIZE, random_neuron).to(device)
            G_data = G(G_noise).detach().clone()

            prob_real = D(selected_real)
            prob_fake = D(G_data)

            D_loss = -torch.mean(torch.log1p(prob_real) + torch.log1p(1. - prob_fake))
            G_loss = torch.mean(torch.log1p(1. - prob_fake))

            opt_D.zero_grad()
            D_loss.backward(retain_graph=True)
            opt_D.step()

        opt_G.zero_grad()
        # Detach gradient computation from the graph
        G_data = G(G_noise).clone().detach() 
        prob_fake = D(G_data)
        G_loss = torch.mean(torch.log1p(1. - prob_fake))
        G_loss.backward()
        opt_G.step()
```

```python
# evaluation
from sklearn.metrics import average_precision_score
from sklearn.metrics import recall_score
test = pd.read_csv('toy_test.csv', header=None)
test = test.values
label = test[:, 2]
label = label.reshape([-1, 1])
test = test[:, :2]
test = torch.tensor(test, device=device)
test_score = 0.0
for iter in range(G_D_num):
    prediction = D_group['D' + str(iter)](test.float()).cpu().detach().numpy()
    test_score = test_score + prediction / G_D_num

alpha = 0.5 * np.max(z_avg)
for i in range(test_score.shape[0]):
    if test_score[i] > alpha:
        test_score[i] = 0
    else:
        test_score[i] = 1

precision = average_precision_score(label, test_score)
recall = recall_score(label, test_score)
F1 = 2 * (precision * recall) / (precision + recall)
print('Precision score: {0:0.2f}'.format(precision))
print('Recall score: {0:0.2f}'.format(recall))
print('F1 score: {0:0.2f}'.format(F1))

h = plt.contourf(x,
                 y,
                 z_avg.reshape(z_avg.shape[0], z_avg.shape[1]),
                 20,
                 cmap='Spectral')
plt.colorbar()

for i in range(test_score.shape[0]):
    if test_score[i] > alpha:
        plt.plot(test.cpu().numpy()[i, 0],
                 test.cpu().numpy()[i, 1],
                 'or',
                 label='Test')
    else:
        plt.plot(test.cpu().numpy()[i, 0],
                 test.cpu().numpy()[i, 1],
                 'ob',
                 label='Test')

plt.show()
```
